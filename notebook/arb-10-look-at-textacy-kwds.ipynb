```python
import pandas as pd
from pathlib import Path
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
infile = Path('../data/full_lim_journal_textrank_05_05_2020/kwds.jsonl')
```

```python
df = pd.read_json(infile, orient='records', lines=True)
```

```python
kwd_exp = df['kwds'].explode()
```

```python
kwd_df = pd.DataFrame.from_items(zip(kwd_exp.index, kwd_exp))
```

```python
kwd_df = kwd_df.T
```

```python
kwd_df.columns = ['keyword', 'score']
```

```python
kwd_df['score'] = kwd_df['score'].astype(np.float)
```

```python
kwd_agg = kwd_df.groupby('keyword').agg({'score': ['count', np.mean]})['score'] #.sort_values('score', ascending=False)
```

```python
pd.set_option('display.max_rows', 100)
```

```python
kwd_agg_lim = kwd_agg.query('count > 3')
```
